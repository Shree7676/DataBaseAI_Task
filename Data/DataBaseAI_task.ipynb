```python
import pandas as pd
```

```python
df = pd.read_json('merged_reviews.json', lines=True)
```

```python
df.drop(['images','asin','parent_asin','user_id'],axis=1,inplace=True)
```

```python
df.groupby('title_y').count()
```

```python
# Sample general product names mapping
product_names = [
    "Black Eye Patch with Tie Band",
    "Fake Toenails Full Cover 100 Pcs",
    "Leather Conditioner 8-Ounce",
    "Baby Silk Sleep Cap",
    "Precision Plunger Bars for Cartridge Grips",
    "Satin Adjustable Sleep Cap (4 Pieces)",
    "Waterproof Tattoo Eyebrow Stickers",
    "Gold Musk EDT 90ml",
    "Tomato Detoxifying Charcoal Cleanser (Pack of 2)",
    "False Eyelash Packaging Box (50 Pcs)"
]

# Matching keywords for each general name
product_keywords = [
    ["eye patch", "black", "tie band"],
    ["toenails", "fake", "full cover", "nail tips"],
    ["leather conditioner", "8-ounce"],
    ["baby", "silk", "sleep cap"],
    ["precision", "plunger bars", "cartridge grips"],
    ["satin", "adjustable", "sleep cap"],
    ["tattoo", "eyebrow", "waterproof"],
    ["gold", "musk", "EDT"],
    ["tomato", "charcoal", "cleanser"],
    ["false", "eyelash", "packaging box"]
]

# Function to match the description with a product name
def assign_product_name(description):
    for name, keywords in zip(product_names, product_keywords):
        if all(keyword.lower() in description.lower() for keyword in keywords):
            return name
    return "Unknown Product"  # In case no match is found

# Apply the function to the merged_df to assign product names
df['product_name'] = df['title_y'].apply(assign_product_name)

# Check the updated DataFrame
df.head()
```

```python
df.drop(['helpful_vote','verified_purchase','timestamp'],axis=1,inplace=True)
```

```python
df.drop('title_y',axis=1,inplace=True)
```

```python
final_data = df.groupby('product_name')[['title_x', 'text']].apply(lambda x: x.values.tolist()).to_dict()
```
